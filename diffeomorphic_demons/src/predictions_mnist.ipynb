```python
warped_image= nib.load('/Users/nastya/Documents/ucu/thesis/diffeomorphic_demons/src/mnist/predictions/pair_65/final_warped_image.nii.gz').get_fdata()
warped_image = np.array(warped_image)

source_image = test_set[65][0]
target_image = test_set[65][1]
```
